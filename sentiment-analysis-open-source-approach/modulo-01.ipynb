Enunciado 1: Supongamos que en una tienda deseas calcular el precio final de un producto aplicando un descuento, pero no quieres repetir la fórmula cada vez que tengas que hacer este cálculo. Puedes usar una función para resolverlo.

```python
def calcular_precio_final(precio_inicial, descuento):
    """
    Calcula el precio final después de aplicar un descuento.
    precio_inicial: Precio original del producto.
    descuento: Porcentaje de descuento (entre 0 y 100).
    """
    precio_final = precio_inicial - (precio_inicial * descuento / 100)
    return precio_final

# Ejemplo práctico
precio_producto = 100  # Precio en dólares
descuento_aplicado = 20  # Porcentaje de descuento

precio_final = calcular_precio_final(precio_producto, descuento_aplicado)
print(f"El precio final después de aplicar el descuento es: ${precio_final}")
```

Enunciado 2: Crear una función que convierta temperaturas de Fahrenheit a Celsius para evitar repetir la fórmula en diferentes partes del programa.

```python
def convertir_a_celsius(fahrenheit):
  """
  Convierte una temperatura de Fahrenheit a Celsius
  """

  celsius = (fahrenheit - 32) * 5 / 9
  return celsius

temperatura_fahrenheit = 69
temperatura_celsius = convertir_a_celsius(temperatura_fahrenheit)
print("Temperatura Fahrenheit igual",temperatura_fahrenheit, "Equivalen a", temperatura_celsius, "Grados Celsius")
```

```python
# Crear DF
import pandas as pd

datos = {
    "Mes": ["Enero", "Febrero", "Marzo", "Abril", "Mayo"],
    "Ventas": [1500, 2300, 1800, 2700, 2200],
    "CLientes_Nuevos": [45, 60, 50, 65, 55],
    "Satisfaccion": [3.8, 4.2, 4.0, 4.5, 4.3]
}

df = pd.DataFrame(datos)
print(df)
```

Grafico: Ventas mensuales usando la base de datos creada en el punto anterior.

```python
import matplotlib.pyplot as plt

plt.plot(df["Mes"], df["Ventas"], marker="o", linestyle="--", color="b", label="Ventas")
plt.title("Tendencia de Ventas Mensuales")
plt.xlabel("Mes")
plt.ylabel("Ventas ($)")
plt.legend()
plt.grid(True)
plt.show()
```

Gráfica: Clientes nuevos por mes

```python
plt.bar(df["Mes"], df["CLientes_Nuevos"], color="orange", alpha=0.8)
plt.title("Clientes Nuevos por Mes")
plt.xlabel("Mes")
plt.ylabel("Clientes Nuevos")
plt.show()
```

Gráfica: Relación entre clientes nuevos y ventas

```python
plt.scatter(df["CLientes_Nuevos"], df["Ventas"], color="red", s=100, alpha=0.7)
plt.title("Relación entre Clientes Nuevos y Ventas")
plt.xlabel("Clientes Nuevos")
plt.ylabel("Ventas ($)")
plt.grid(True)
plt.show()
```
